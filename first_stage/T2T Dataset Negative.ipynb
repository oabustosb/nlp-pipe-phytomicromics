## Imports

```python
import numpy as np
import pandas as pd
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from transformers import pipeline
```

## Database

```python
database = pd.read_excel('DatasetNegative.xlsx')
database
```

```python
database = database.dropna(subset="ABSTRACT")
database.drop_duplicates(subset="ABSTRACT", inplace=True)
database
```

```python
database.sort_values(by="TITLE", inplace=True)
database.reset_index(drop=True, inplace=True)
database
```

```python
abstracts = database['ABSTRACT'].to_list()
```

## T5

### Pregunta 1

```python
responses_1_1 = []
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

for text in abstracts:
    input = f"question: There is a reference to antimicrobial, antibacterial or antiviral activity? Say yes or no, context: {text}"
    input_ids = tokenizer(input, return_tensors="pt").input_ids
    try:
        outputs = model.generate(input_ids)
        responses_1_1.append(tokenizer.decode(outputs[0]))
    except:
        responses_1_1.append('No')
```

```python
print(responses_1_1)
```

```python
print(np.unique(responses_1_1, return_counts=True))
```
